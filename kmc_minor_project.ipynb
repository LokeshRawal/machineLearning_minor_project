## Description:
This dataset contains simulated data representing customer purchase behavior. It includes various features such as age, gender, income, education, region, loyalty status, purchase frequency, purchase amount, product category, promotion usage, and satisfaction score.

## File Information
1. File Format: CSV
2. Number of Rows: 100000
3. Number of Columns: 12

## Column Descriptors:
1. age: Age of the customer.
2. gender: Gender of the customer (0 for Male, 1 for female).
3. income: Annual income of the customer.
4. education: Education level of the customer.
5. region: Region where the customer resides.
6. loyalty_status: Loyalty status of the customer.
7. purchase_frequency: Frequency of purchase made by the customer
8. purchase_amount: Amount spent by the customer in each purpose.
9. product_category: Category of the purchased product.
10. promotion_usage: Indicates whether the customer used promotional offers (0 for No, 1 for Yes).
11. satisfaction_score: Satisfaction score of the customer. customer.

```python
def univariate_barplot(data, xlabel, ylabel="Frequency", xlabel_rotation=0):
    # Set font size
    sns.set(font_scale=1.4)

    # Get value counts
    value_counts = data[xlabel].value_counts()
    ax = value_counts.plot.bar(
        # Set a categorical color palette
        color=sns.color_palette("Set2"),
        # Set a figure size
        figsize=(10, 10),
        # Rotate the x-axis labels for proper orientation
        rot=xlabel_rotation,
    )

    total = value_counts.sum()
    # Add percentage label for each plot
    for plot in ax.patches:
        height = plot.get_height()
        ax.annotate(
            # Percentage label text
            f"{height / total * 100:.1f}%",
            xy=(
                # Percentage label x co-ordinate
                plot.get_x() + plot.get_width() / 2.,
                # Percentage label y co-ordinate
                height,
            ),
            # Percentage label horizontal align
            ha="center",

        )

    ax.set_title(f"Distribution of {xlabel} in the dataset", y=1.02)
    ax.set_xlabel(xlabel, labelpad=14)
    ax.set_ylabel(ylabel, labelpad=14)
    
```

```python
def fit_label_encoder(feature):
    le = LabelEncoder()
    le.fit(feature)
    return le
```

```python
def one_hot_encoder(data, feature_name, column_to_drop=None):
    modified_data = data.copy()
    feature = modified_data[feature_name]

    if column_to_drop is None or column_to_drop not in feature.unique():
        # Drop the column with least number of data points
        column_to_drop = feature.value_counts().index[-1]

    dummies = pd.get_dummies(feature)
    modified_data[dummies.columns] = dummies

    # Convert dummy columns to integers (1 or 0)    
    # modified_data[dummies.columns] = modified_data[dummies.columns].astype(int)

    return modified_data.drop([column_to_drop, feature_name], axis=1)
```

### Choosing the correct encoder

```python
feature_columns = categorical_features + numerical_features
features = df[feature_columns].copy()

gender_encoder = fit_label_encoder(feature=features.gender)
features["gender"] = gender_encoder.transform(features["gender"])

region_encoder = fit_label_encoder(feature=features.region)
features["region"] = region_encoder.transform(features["region"])

product_category_encoder = fit_label_encoder(feature=features.product_category)
features["product_category"] = product_category_encoder.transform(features["product_category"])

features = one_hot_encoder(features, "education", column_to_drop="Bachelor")
features = one_hot_encoder(features, "loyalty_status", column_to_drop="Gold")
features = one_hot_encoder(features, "purchase_frequency", column_to_drop="frequent")

features.head(10)
```

## InterQuartile Range

```python
def get_iqr(feature):
    quantile1, quantile3 = np.percentile(feature, [25, 75])
    iqr_value = quantile3 - quantile1
    lower_bound = quantile1 - 1.5 * iqr_value
    upper_bound = quantile3 + 1.5 * iqr_value
    return lower_bound, upper_bound
```

```python
lower, upper = get_iqr(features.age)
outlier_condition = (features.age < lower) | (features.age > upper)
features[outlier_condition] = features[outlier_condition].mean()
```
